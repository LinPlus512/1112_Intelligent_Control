```python
import torch
import json
from tabulate import tabulate
```

```python
# 讀取JSON文件
file_path = './youbike_immediate.json'
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)
print(len(data))
print(data[0])
```

```python
import json
import pandas as pd

# 取出有用的欄位
df = pd.DataFrame(data, columns=['sno', 'tot', 'sbi', 'mday', 'bemp'])

# 轉換時間欄位為 datetime 格式

df['mday'] = pd.to_datetime(df['mday'], format='%Y-%m-%d %H:%M:%S')
df['mday'] = df['mday'].apply(lambda x: pd.to_datetime(x, unit='s'))

# 設定車輛編號為索引
df = df.set_index('sno')

# 將資料集存成 CSV 檔案
df.to_csv('./dataset.csv')

```

```python
# 將字典轉換為列表
table_data = [[d[key] for key in d] for d in data]

# 印出表格
headers = list(data[0].keys())
print(tabulate(table_data, headers=headers))
```
