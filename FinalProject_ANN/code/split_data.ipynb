## Add sbi_30min
The number of stations after adding 30 minutes.

```python
# read date
data = pd.read_csv('split_data_test.csv')
# data transform
ann_data = data[['sno', 'tot', 'sbi']]
```

```python
print(len(ann_data))
print(len(ann_data.iloc[:len(ann_data) - 1279 * 30]['sno'].values))
```

```python
min = 30 # Customized data after a specific number of minutes.

sno_30min = (ann_data.iloc[0 + 1279 * min:]['sno'].values)
sbi_30min = (ann_data.iloc[0 + 1279 * min:]['sbi'].values)
print(len(sno_30min))
```

```python
datasets = pd.DataFrame()
datasets['sno'] = ann_data.iloc[:len(ann_data) - 1279 * min]['sno'].values
datasets['tot'] = ann_data.iloc[:len(ann_data) - 1279 * min]['tot'].values
datasets['sbi'] = ann_data.iloc[:len(ann_data) - 1279 * min]['sbi'].values
# add column
datasets['sno_30min'] = sno_30min
datasets['sbi_30min'] = sbi_30min
# save
datasets.to_csv('ann_datasets_test.csv', index=False)
```
