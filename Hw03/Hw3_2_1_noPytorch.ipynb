# Realize HW03 question 2-1 with no Pytorch

## import package

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

## define parameter

```python
# inputs
x = np.array([[-1., -1.],
              [-1.,  1.],
              [ 1., -1.],
              [ 1.,  1.]])
# desire output
Yd = np.array([[0.],
               [1.],
               [1.],
               [0.]])
# weighting matrix
w = np.array([1., -1.])

# learning rate
lr = 0.1

#threshold
theta = 0

# epochs
epochs = 10000
```

## calculate train process

```python
# Loss function
def MSELoss(e):
    return np.sum(np.square(e))
```

```python
loss = []
for i in tqdm(range(epochs)):
    # forward pass
    Yp = np.where(np.dot(x, w) > theta, 1., 0.)
    # error calculation
    error = Yd - Yp.reshape(Yd.shape)
    # loss.append(MSELoss(error))
    loss.append(error.tolist())
    # update weights
    w += lr * np.dot(x.T, error).reshape(w.shape)
```

## Plot Loss curve

```python
# plot loss curve with stride
plt.plot(np.array(loss).flatten()[:50])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()
```
